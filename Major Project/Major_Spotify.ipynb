```python
def top_correlations(corr_matrix, top_n=10):
    cols = corr_matrix.columns
    pairs = []
    for i in range(len(cols)):
        for j in range(i+1, len(cols)):
            pairs.append((cols[i], cols[j], corr_matrix.iloc[i, j]))
    # sort by absolute correlation descending
    pairs_sorted = sorted(pairs, key=lambda x: -abs(x[2]))
    return pairs_sorted[:top_n]

top10 = top_correlations(pearson_corr, top_n=12)
print("Top correlated feature pairs (Pearson, by absolute value):")
for a, b, v in top10:
    print(f"{a} <-> {b} : corr = {v:.3f}")

# Also show the most strongly negative pairs
neg_pairs = sorted([(a,b,v) for a,b,v in top10 if v < 0], key=lambda x: x[2])  # ascending (more negative first)
if neg_pairs:
    print("\nStrong negative correlations in top list:")
    for a,b,v in neg_pairs:
        print(f"{a} <-> {b} : corr = {v:.3f}")
```

```python
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize
from scipy.sparse import csr_matrix
from sklearn.cluster import KMeans

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
random.seed(RANDOM_STATE)
plt.rcParams["figure.figsize"] = (7,4)
```

```python

print("Raw shape:", df.shape)

# Drop rows missing essentials
needed = [c for c in ("track_id","track_name","track_artist","playlist_id") if c in df.columns]
df = df.dropna(subset=needed).reset_index(drop=True)
print("After dropping missing essentials:", df.shape)

# create duration_min (if exists)
if "duration_ms" in df.columns:
    df["duration_min"] = df["duration_ms"] / 60000.0

# tracks-level unique table (one row per track)
if "track_id" in df.columns:
    df_tracks = df.drop_duplicates(subset=["track_id"]).reset_index(drop=True)
else:
    df_tracks = df.drop_duplicates(subset=["track_name","track_artist"]).reset_index(drop=True)

print("Unique tracks:", len(df_tracks))
```

```python
feature_cols = [c for c in ["danceability","energy","loudness","speechiness",
                            "acousticness","instrumentalness","liveness",
                            "valence","tempo","duration_min","track_popularity"]
                if c in df_tracks.columns]
print("Features used:", feature_cols)

# Fill NaNs with medians (simple)
for c in feature_cols:
    if df_tracks[c].isnull().any():
        df_tracks[c] = df_tracks[c].fillna(df_tracks[c].median())

# Scale then normalize (standard practice before cosine similarity)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_tracks[feature_cols].values)
normalizer = Normalizer()
X_norm = normalizer.fit_transform(X_scaled)  # rows are unit vectors

# Map track_id <-> index
track_ids = df_tracks["track_id"].astype(str).tolist()
track_id_to_idx = {tid: i for i, tid in enumerate(track_ids)}
idx_to_track_id = {i: tid for tid, i in track_id_to_idx.items()}

# Add index column to df_tracks for convenience
df_tracks["track_idx"] = df_tracks.index
```

```python
playlist_ids = df["playlist_id"].astype(str).unique().tolist()
playlist_id_to_idx = {pid: i for i, pid in enumerate(playlist_ids)}
idx_to_playlist_id = {i: pid for pid, i in playlist_id_to_idx.items()}

n_playlists = len(playlist_ids)
n_tracks = len(track_ids)
print("Number of playlists:", n_playlists, "number of unique tracks:", n_tracks)

# Build rows, cols, data to create a CSR matrix
rows = []
cols = []
data_vals = []
for _, row in df.iterrows():
    pid = str(row["playlist_id"])
    tid = str(row["track_id"])
    if tid in track_id_to_idx:   # safety check
        rows.append(playlist_id_to_idx[pid])
        cols.append(track_id_to_idx[tid])
        data_vals.append(1)

# Create csr matrix (playlist x track)
P_csr = csr_matrix((data_vals, (rows, cols)), shape=(n_playlists, n_tracks), dtype=np.int32)

# Normalize playlist rows (for cosine comparisons later)
P_norm = normalize(P_csr, axis=1)  # sparse normalized
# Also normalized track vectors (track x playlist) for item-based ops
T_norm = normalize(P_csr.T, axis=1)  # shape (n_tracks x n_playlists)
```

```python
def recommend_content_for_playlist(playlist_id, top_n=10, exclude_track_ids=None):
    """
    playlist_id: as present in df['playlist_id'] (string or numeric)
    exclude_track_ids: list of track_id values (strings) to exclude from centroid calculation (e.g., holdout)
    returns DataFrame of recommendations
    """
    pid = str(playlist_id)
    if pid not in playlist_id_to_idx:
        print("Playlist not found:", pid)
        return None

    # tracks in playlist
    plist_tracks = df[df["playlist_id"].astype(str) == pid]["track_id"].astype(str).unique().tolist()
    if exclude_track_ids:
        plist_tracks = [t for t in plist_tracks if str(t) not in set(map(str, exclude_track_ids))]
    if len(plist_tracks) == 0:
        print("No tracks left in playlist after excluding. Cannot build centroid.")
        return None

    idxs = [track_id_to_idx[str(t)] for t in plist_tracks if str(t) in track_id_to_idx]
    if len(idxs) == 0:
        print("No matching track indices for playlist tracks.")
        return None

    # compute centroid in normalized feature space
    centroid = X_norm[idxs].mean(axis=0)
    norm = np.linalg.norm(centroid)
    if norm == 0:
        centroid = centroid
    else:
        centroid = centroid / norm

    # score all tracks by dot product (cosine similarity)
    scores = X_norm.dot(centroid)  # shape (n_tracks,)
    scores = np.array(scores).ravel()

    # exclude tracks already in playlist (we want to recommend new tracks), but do not exclude holdout (we will control this externally)
    existing = set([str(x) for x in df[df["playlist_id"].astype(str) == pid]["track_id"].astype(str).unique().tolist()])
    # We'll not remove holdout here; evaluation will pass exclude list as needed

    # build DataFrame with scores
    rec_df = df_tracks[["track_id","track_name","track_artist","track_popularity"]].copy()
    rec_df["score"] = scores
    # filter out existing tracks if you want only new recommendations:
    rec_df = rec_df[~rec_df["track_id"].astype(str).isin(existing)]
    rec_df = rec_df.sort_values("score", ascending=False).head(top_n).reset_index(drop=True)
    return rec_df
```

```python
from scipy.sparse import csr_matrix

def recommend_by_similar_playlists(playlist_id, top_n=10, exclude_track_ids=None, n_neighbors=20):
    pid = str(playlist_id)
    if pid not in playlist_id_to_idx:
        print("Playlist not found:", pid)
        return None
    pidx = playlist_id_to_idx[pid]

    # Build target playlist vector (binary) but exclude certain tracks from its vector if given
    playlist_row = P_csr.getrow(pidx).toarray().ravel().astype(np.float32)  # dense for this playlist only
    if exclude_track_ids:
        for t in exclude_track_ids:
            if str(t) in track_id_to_idx:
                playlist_row[track_id_to_idx[str(t)]] = 0.0

    # Normalize target vector
    if playlist_row.sum() == 0:
        print("Playlist vector is empty after excluding tracks.")
        return None
    p_vec_norm = playlist_row / np.linalg.norm(playlist_row)

    # Compute cosine similarity between target playlist and all playlists (use P_norm which is normalized)
    # sim = P_norm * p_vec_norm (since P_norm rows are normalized)
    sim_scores = P_norm.dot(p_vec_norm)  # shape (n_playlists,)
    sim_scores = np.array(sim_scores).ravel()

    # ignore the playlist itself
    sim_scores[pidx] = 0.0

    # pick top neighbor playlist indices
    neighbor_idxs = np.argsort(sim_scores)[::-1][:n_neighbors]
    neighbor_sims = sim_scores[neighbor_idxs]

    # aggregate neighbor tracks weighted by similarity
    # get neighbor x tracks submatrix
    neighbor_matrix = P_csr[neighbor_idxs]  # shape (n_neighbors x n_tracks)
    # weighted sum: weights = neighbor_sims
    weights = neighbor_sims.reshape(-1,1)
    track_scores = (neighbor_matrix.multiply(weights)).sum(axis=0).A1  # shape (n_tracks,)

    # remove tracks already in the original playlist (we want recommendations)
    existing_tracks = set(df[df["playlist_id"].astype(str) == pid]["track_id"].astype(str).tolist())
    # if exclude_track_ids was used for holdout, we want holdout to be a candidate (so do not add holdout to existing set)
    if exclude_track_ids:
        existing_tracks = existing_tracks - set([str(x) for x in exclude_track_ids])

    # create recommendation DataFrame
    rec_df = df_tracks[["track_id","track_name","track_artist","track_popularity"]].copy()
    rec_df["score"] = track_scores
    rec_df = rec_df[~rec_df["track_id"].astype(str).isin(existing_tracks)]  # exclude existing remaining tracks
    rec_df = rec_df.sort_values("score", ascending=False).head(top_n).reset_index(drop=True)
    return rec_df
```

```python
def recommend_ensemble(playlist_id, top_n=10, exclude_track_ids=None, alpha=0.6):
    """
    alpha: weight for content-based, (1-alpha) for playlist-collab
    """
    # get content scores (we'll compute full ranking instead of already filtered top-n)
    # compute centroid from remaining tracks (exclude exclude_track_ids)
    pid = str(playlist_id)
    plist_tracks = df[df["playlist_id"].astype(str) == pid]["track_id"].astype(str).unique().tolist()
    remaining = [t for t in plist_tracks if exclude_track_ids is None or str(t) not in set(map(str, exclude_track_ids))]
    if len(remaining) == 0:
        print("No remaining tracks to build centroid/playlist vector.")
        return None

    idxs = [track_id_to_idx[str(t)] for t in remaining if str(t) in track_id_to_idx]
    centroid = X_norm[idxs].mean(axis=0)
    norm = np.linalg.norm(centroid)
    centroid = centroid / norm if norm>0 else centroid
    content_scores = X_norm.dot(centroid).ravel()

    # get playlist-collab track scores (as array)
    # reuse recommend_by_similar_playlists but we need full score vector, so we'll compute same steps but return track_scores
    pidx = playlist_id_to_idx[pid]
    playlist_row = P_csr.getrow(pidx).toarray().ravel().astype(np.float32)
    # exclude specified tracks from the playlist vector (so they are predicted from neighbors)
    if exclude_track_ids:
        for t in exclude_track_ids:
            if str(t) in track_id_to_idx:
                playlist_row[track_id_to_idx[str(t)]] = 0.0
    p_vec_norm = playlist_row / np.linalg.norm(playlist_row)
    sim_scores = P_norm.dot(p_vec_norm).ravel()
    sim_scores[pidx] = 0.0
    neighbor_idxs = np.argsort(sim_scores)[::-1][:20]
    neighbor_sims = sim_scores[neighbor_idxs]
    neighbor_matrix = P_csr[neighbor_idxs]
    track_scores_collab = (neighbor_matrix.multiply(neighbor_sims.reshape(-1,1))).sum(axis=0).A1

    # normalize both score arrays to 0..1 to combine safely
    def _norm_arr(a):
        a = np.array(a).astype(np.float64)
        mn, mx = a.min(), a.max()
        if mx - mn < 1e-9:
            return a - mn
        return (a - mn) / (mx - mn)

    cs = _norm_arr(content_scores)
    ps = _norm_arr(track_scores_collab)
    combined = alpha * cs + (1-alpha) * ps

    # exclude remaining playlist tracks from candidates (but allow holdout if exclude list given)
    existing = set(df[df["playlist_id"].astype(str) == pid]["track_id"].astype(str).tolist())
    if exclude_track_ids:
        existing = existing - set([str(x) for x in exclude_track_ids])

    rec_df = df_tracks[["track_id","track_name","track_artist","track_popularity"]].copy()
    rec_df["score"] = combined
    rec_df = rec_df[~rec_df["track_id"].astype(str).isin(existing)]
    rec_df = rec_df.sort_values("score", ascending=False).head(top_n).reset_index(drop=True)
    return rec_df
```

```python
from tqdm import tqdm

def evaluate_recommender_on_playlists(recommender_fn, df, n_samples=200, K=10, min_tracks=3):
    """
    recommender_fn: function(playlist_id, top_n=K, exclude_track_ids=[holdout]) -> DataFrame of recs
    This function performs leave-one-out: pick one holdout track per playlist, compute recommendations using remaining tracks,
    and check whether holdout appears in top-K. Returns hit-rate and MRR.
    """
    playlist_candidates = []
    grouped = df.groupby("playlist_id")["track_id"].nunique()
    for pid, cnt in grouped.items():
        if cnt >= min_tracks:
            playlist_candidates.append(pid)
    if not playlist_candidates:
        print("No playlists with >= min_tracks")
        return None

    sample_pids = random.sample(playlist_candidates, min(n_samples, len(playlist_candidates)))

    hits = 0
    rr_sum = 0.0
    tested = 0

    for pid in tqdm(sample_pids, desc="Evaluating"):
        # choose a random holdout
        tracks = df[df["playlist_id"] == pid]["track_id"].astype(str).unique().tolist()
        holdout = random.choice(tracks)
        # remaining tracks
        remaining = [t for t in tracks if t != holdout]
        if len(remaining) == 0:
            continue

        # get recommendations while building model using remaining (pass exclude_track_ids so centroid/playlist vector excludes holdout)
        recs = recommender_fn(pid, top_n=K, exclude_track_ids=[holdout])
        if recs is None or recs.shape[0] == 0:
            continue
        rec_track_ids = recs["track_id"].astype(str).tolist()
        tested += 1
        if str(holdout) in rec_track_ids:
            hits += 1
            rank = rec_track_ids.index(str(holdout)) + 1
            rr_sum += 1.0 / rank
        else:
            rr_sum += 0.0

    hr = hits / tested if tested > 0 else 0.0
    mrr = rr_sum / tested if tested > 0 else 0.0
    return {"tested": tested, "HR@K": hr, "MRR@K": mrr}

# Run quick evaluation on a sample:
print("Evaluating content-based recommender...")
res_content = evaluate_recommender_on_playlists(recommend_content_for_playlist, df, n_samples=200, K=10, min_tracks=3)
print(res_content)

print("Evaluating playlist-collab recommender...")
res_collab = evaluate_recommender_on_playlists(recommend_by_similar_playlists, df, n_samples=200, K=10, min_tracks=3)
print(res_collab)

print("Evaluating ensemble recommender (alpha=0.6)...")
res_ensemble = evaluate_recommender_on_playlists(lambda pid, top_n, exclude_track_ids: recommend_ensemble(pid, top_n=top_n, exclude_track_ids=exclude_track_ids, alpha=0.6),
                                                df, n_samples=200, K=10, min_tracks=3)
print(res_ensemble)
```
